```python
new_york_data_df.insert(new_york_data_df.shape[1],'date',new_york_data_df['Registration Valid Date'].str.split('-', expand = True)[0])
```

```python
new_york_data_df.insert(new_york_data_df.shape[-1],'make',new_york_data_df['Vehicle Name'].str.split(' ', expand = True)[0])
```

```python
new_york_data_df
```

```python
ny_df_h2 = new_york_data_df[['date','make']]
```

```python
ny_df_h2
```

```python
ny_df_h2_countyears = ny_df_h2[pd.notnull(ny_df_h2['date'])].groupby(by='date').count()
ny_df_h2_countyears.columns = ['Count']
ny_df_h2_countyears
```

From the year count chart, we chose the recent 5 years except 2022, because it is incomplete. So we chose 2017-2021 years.

```python
year_list = ['2017', '2018', '2019', '2020', '2021']
for year in year_list:
    ny_df_h2_year = ny_df_h2.loc[ny_df_h2['date'] == year]
    ny_df_h2_year_count = ny_df_h2_year[pd.notnull(ny_df_h2_year['make'])].groupby(by='make').count()
    ny_df_h2_year_count.columns = ['Count' + ' ' + str(year)]
    print(ny_df_h2_year_count)
    ny_df_h2_year_count.plot.pie(subplots=True, figsize=(15, 15), autopct='%.1f',
                                 title=str(year) + 'Pie Chart of Different Brands of Car in ' + str(year))
    ny_df_h2_year_count.plot.bar(figsize=(20, 10),
                                 title=str(year) + 'Bar Chart of Different Brands of Car in ' + str(year))
    print()
```
